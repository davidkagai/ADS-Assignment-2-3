```python
data["DistanceFromHome_title"] = pd.cut(data.DistanceFromHome,[-1,5,10,20,30], labels=['Very near','near','far','very far'])
```

```python
data.groupby('DistanceFromHome_title')['Attrition']\
    .value_counts()\
    .unstack(level=1)\
    .plot.bar(stacked=True)
pd.crosstab(data.DistanceFromHome_title, data.Attrition, margins=True, normalize='index').round(2).style.background_gradient(cmap='spring')
```

```python
data.groupby('JobRole')['DistanceFromHome_title']\
    .value_counts()\
    .unstack(level=1)\
    .plot.bar(stacked=True, figsize=(7,6))
```

```python
data["NumCompaniesWorked_grp"] = pd.cut(data.NumCompaniesWorked,[-1,4,9], labels=['Less than 5','More than 5'])


data.groupby('NumCompaniesWorked_grp')['Attrition']\
    .value_counts()\
    .unstack(level=1)\
    .plot.bar(stacked=True)

pd.crosstab(data.NumCompaniesWorked_grp, data.Attrition, margins=True, normalize='index').round(2).style.background_gradient(cmap='spring')
```

Employees who have worked in more than 5 companies are more likely to leave the organisation

What we can conclude from the plots above is

1. The employees who turned over are generally younger than the stayed employees.



2. Employees from job roles with higher average age and requires to travel rarely are relatively most likely to turn over.

The underlying mechanism can be:

1. Younger employees has less financial stress.

2. Employees that travel more frequently might like to travel, while some employees who left because of rare travels did not enjoy travel at all.
